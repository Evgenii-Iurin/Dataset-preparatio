```python
def extract_lable(FILE_PATH, FILE_NAME, EXTRACT_PATH):
    non_annotation_images = 0;
    annotation_images = 0;
    
    """ This function count the non-annotated and annotated images in one TFRecord file
                
        Parameters
        ----------
        FILE_PATH - path file from which information is extracted
        FILE_NAME - TFRecord file name
        EXTRACT_PATH - folder path for extracting data
        
        Return
        ----------
        Number of annotated images and non-annotated images in one TFRecord file
    
    """
    
    """ Read the TFRecord format file from folder """
    tfr_record_file = tf.data.TFRecordDataset(FILE_PATH+FILE_NAME, compression_type='')

    
    for data in tfr_record_file:
        
        """ Read one frame  """
        frame = open_dataset.Frame()
        frame.ParseFromString(bytearray(data.numpy()))
        
        
        
        """ This is itarator each image in one frame """
        for camera_labels in frame.camera_labels:
            if not len(camera_labels.labels):
                non_annotation_images +=1
            elif len(camera_labels.labels):
                annotation_images +=1
    
    
    return non_annotation_images, annotation_images



def main():
    FOLDER_PATH = '/nfs-home/Prepare_Waymo_Dataset/Waymo_dataset/train/'
    EXTRACT_PATH = '/nfs-home/Prepare_Waymo_Dataset/Waymo_dataset_extracted_data/train/'
    all_non_ann_img = 0;
    all_ann_img = 0;
    
    for FILENAME in tqdm.tqdm(os.listdir(FOLDER_PATH)):
        if FILENAME.endswith('.tfrecord'):
                non_ann_img, ann_img = extract_lable(FOLDER_PATH, FILENAME, EXTRACT_PATH)
        all_non_ann_img += non_ann_img
        all_ann_img += ann_img
    print(f'Non-annotation images: {all_non_ann_img}\nAnnotation images: {all_ann_img}')
    
if __name__ == '__main__':
    main() 
```
